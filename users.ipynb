```python
def FirstName(x):
    name = x.split(" ")[0]
    return name.capitalize()

dfZ303['firstname'] = dfZ303['z303_name'].apply(FirstName)
```

```python
def Middle_name(x):
    preNames = ['DE', 'DA', 'DAS', 'DO', 'DOS']
    name = x.split()
    name = name[1:-1]
    
    name = [i.capitalize() if i not in preNames else i.lower() for i in name ]
    return " ".join(name)

dfZ303['middle_name'] = dfZ303['z303_name'].apply(Middle_name) 
```

```python
def Surname(x):
    name = x.split(" ")[-1]
    return name.capitalize()

dfZ303['surname'] = dfZ303['z303_name'].apply(Surname)
```

```python
for index, row in dfZ303[filtro].iterrows():
    # firstname	middle_name	surname
    if type(row.middle_name) != float:
        name = f'{row.firstname.upper()} {row.middle_name.upper()} {row.surname.upper()}'
    else:
        name = f'{row.firstname.upper()} {row.surname.upper()}'

    address = row.address.replace(name, '')
    print(address)
    dfZ303.at[index, 'address'] = address
    
```

```python
import re
def RemoveEspacoEmBranco(x):
    if type(x) == float or x == None:
        return x
    else:
        return re.sub(r"\s+", " ", x).lstrip()
    
dfZ303['address'] = dfZ303['address'].apply(RemoveEspacoEmBranco)
```

```python
def RemoveEnderecoAutomatico(x):
    s = 'gerado automaticamente pelo sistema'
    if x == None or type(x) == float:
        return x
    elif s in x:
        return None
    else: 
        return x
    
dfZ303['address'] = dfZ303['address'].apply(RemoveEnderecoAutomatico)
```
